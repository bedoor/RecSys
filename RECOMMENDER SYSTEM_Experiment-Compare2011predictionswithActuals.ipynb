```python
# Predict the unknown ratings through the dot product of the latent features for users and items 
def prediction(P,Q):
    return np.dot(P.T,Q)

def predictionBias(P,Q,mu, bu, bi):
    return mu + bu + bi + np.dot(P.T,Q)

def predictionBias2(P,Q,mu, bu, bi):
    return mu + np.vstack(bu) + bi + np.dot(P.T,Q)

# Calculate the RMSE
def rmse(I,R,Q,P):
    return np.sqrt(np.sum((I * (R - prediction(P,Q)))**2)/len(R[R > 0]))

def rmseBias(I,R,Q,P,mu,bu,bi):
    #return np.sqrt(np.sum((I * (R - predictionBias(P,Q,mu, bu, bi)))**2)/len(R[R > 0]))
     
    return np.sqrt(np.sum((I * (R - mu - bi - np.vstack(bu) - prediction(P,Q)))**2)/len(R[R > 0]))
```

## Similarity measure 1: Giving weight to each rank

```python
counter = 0
T = 0
F = 0


for aindx in temp['Aindex'].unique():
    
    Vinfo = temp[temp['Aindex']==aindx][['VID']].merge(Vindex, on='VID')
    
    actual = pd.DataFrame()
    actual['Vindex'] =  Vinfo[['Vindex']] 
    
    #give top 20 weights, rest is zeros
    predicted = pd.DataFrame()
    predicted['Vindex'] = R_hat.ix[aindx].sort_values(ascending=False).head(n=100).index.values
    predicted['SimValues'] = np.arange(1,0,-0.01)
    
    res = actual.merge(predicted, on='Vindex')
    
    if res.empty:
        F=F+1
    else:
        T=T+1
        result.set_value(aindx, 'SimIndex', res.SimValues.sum())
    
    #counter = counter+1

print "T", T
print "F", F
#result
```

## Similarity measure 2: Using Probability

```python
import math


for aindx in temp['Aindex'].unique():
   
    Vinfo = temp[temp['Aindex']==aindx][['VID']].merge(Vindex, on='VID')
    
    actual =  Vinfo['Vindex'].drop_duplicates().tolist() #drop duplicates
    
    Vnum = len(actual)

    predicted = R_hat.ix[aindx].sort_values(ascending=False)
    
    predicted2 = predicted[predicted>0]
    probs = predicted2[actual]/predicted2.sum()

    total = 0

    for i in actual:
        total = total + math.log(probs[i],2)

    result.set_value(aindx, 'Perplexity', 1/(math.pow(2,(-total/Vnum))))

#print result   
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats
%matplotlib inline

def draw_zoomed(year, result2):
    sns.set_style("darkgrid")
    sns.set_context("poster")

    f, (ax1) = plt.subplots(figsize=(6,6), ncols=1, sharey=True)
    f.suptitle("Average Citations of Authors against their \n% of Similarity between Prediction and Reality")
    plt.subplots_adjust(top=0.85)

    x_jit = 0.0
    y_jit = 0.0

    ax1.set_xlim([0,1.01])
    ax1.set_ylim([0,50])
    sns.regplot(result2['PercSim'], result2['Citations'], ax=ax1, marker='o', color='#F05C5C', scatter=True, x_jitter=x_jit, y_jitter=y_jit, x_ci=99,\
                scatter_kws={'c':'#89B789'})
    ax1.set_ylabel("Average Number of Citations")
    ax1.set_xlabel("% of Similarity between Recommended and Actual")

    plt.savefig("../data/images/RS/PercSimilarityvsAvgCitations"+str(year)+"_zoomed.jpg")
```
